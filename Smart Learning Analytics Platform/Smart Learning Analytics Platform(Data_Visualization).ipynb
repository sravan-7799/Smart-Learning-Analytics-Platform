# **Data preprocessing**

Add New Feature

```python
# Define the passing threshold
passing_threshold = 70

# Update the 'Grade' column based on the passing threshold
df['Grade'] = df['Exam_Score'].apply(lambda x: 'Pass' if x >= passing_threshold else 'Fail')

df.head()
```

```python
#understanding the binary categorical features
binary_categorical_features = [col for col in df.select_dtypes(include='object') if df[col].nunique() == 2]
# Display the binary categorical features
print("Binary Categorical Features:", binary_categorical_features)
```

```python
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder object
le = LabelEncoder()
binary_categorical_features = [col for col in df.select_dtypes(include='object') if df[col].nunique() == 2]

# Apply Label Encoding to each binary feature
for feature in binary_categorical_features:
    df[feature] = le.fit_transform(df[feature])

df.head()
```

```python
#understanding the non binary categorical features
non_binary_categorical_features = [col for col in df.select_dtypes(include='object') if df[col].nunique() > 2]
print("Non-Binary Categorical Features:", non_binary_categorical_features)
```

```python
from sklearn.preprocessing import OrdinalEncoder

# Create an OrdinalEncoder object
encoder = OrdinalEncoder()
encoder.fit(df[non_binary_categorical_features])
df[non_binary_categorical_features] = encoder.transform(df[non_binary_categorical_features])

# Display the updated DataFrame
df.head()
```

```python
#get a copy of encoded data and check the null values
df_encoded = df.copy()
df_encoded.isnull().sum()
```

Scatterplots

```python
#Scatterplot
def create_scatterplot(df, x_feature, y_feature):

    if x_feature in df.columns and y_feature in df.columns:
        plt.figure(figsize=(8, 6))
        plt.scatter(df[x_feature], df[y_feature], alpha=0.5)
        plt.title(f'{y_feature} vs. {x_feature}')
        plt.xlabel(x_feature)
        plt.ylabel(y_feature)
        plt.grid(True)
        plt.show()
    else:
        print(f"Either '{x_feature}' or '{y_feature}' column not found in the DataFrame.")


feature_combinations = [
    ('Hours_Studied', 'Exam_Score'),
    ('Previous_Scores', 'Exam_Score'),
    ('Attendance', 'Exam_Score'),
    ('Sleep_Hours', 'Exam_Score'),

]

for x_feature, y_feature in feature_combinations:
    create_scatterplot(df_encoded, x_feature, y_feature)
```
